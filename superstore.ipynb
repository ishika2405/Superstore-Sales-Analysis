```python
import pandas as pd

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('/content/Sample - Superstore.xlsx')

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
df.info()
df.isnull().sum()
df.duplicated().sum()

```

```python
df.drop_duplicates(inplace=True)

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df["Profit Margin"] = df["Profit"] / df["Sales"]
df["Delivery Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month_name()
```

```python
df.to_csv("cleaned_superstore.csv", index=False)
```

```python
# Sales by category
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
 # Profit by category
df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

```

```python
# Region sales
df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

```

```python
# Top 10 customers
df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

```

```python
# Loss products
df.groupby("Product Name")["Profit"].sum().sort_values().head(10)
```

```python

# Discount impact
df.groupby("Discount")[["Sales","Profit"]].sum()
```
